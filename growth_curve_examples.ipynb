```python
import cobra
import numpy as np
import matplotlib
import scipy
```

```python
model_paths = {'L_lactis': './models/L_lactis/L_lactis.xml',
                'L_plantarum': './models/L_plantarum/L_plantarum.xml',
                'S_cerevisiae': './models/S_cerevisiae/12918_2012_1082_MOESM2_ESM.XML'}
```

```python
%%capture
l_lactis = cobra.io.read_sbml_model(model_paths['L_lactis'])
l_plant_model = cobra.io.read_sbml_model(model_paths['L_plantarum'])
yeast_model = cobra.io.read_sbml_model(model_paths['S_cerevisiae'])
```

```python
# cobra.io.validate_sbml_model(model_paths['S_cerevisiae'])
```

```python
l_lactis_opt = l_lactis.optimize()
l_plant_opt = l_plant_model.optimize()
```

```python
l_lac_df = l_lactis_opt.to_frame()
l_plant_df = l_plant_opt.to_frame()
```

```python
print(l_plant_df)
```

```python
l_plant_df.index.name = 'metabolite'
l_lac_df.index.name = 'metabolite'

l_lac_df.reset_index(inplace=True)
l_plant_df.reset_index(inplace=True)
```

```python
merge_df =  l_lac_df.merge(l_plant_df, on='metabolite', suffixes=['_l_lac', '_l_plant'])
```

```python
lac_to_plant_feeding = merge_df.loc[ (merge_df['fluxes_l_lac'] > 0) & (merge_df['fluxes_l_plant'] < 0) ]
plant_to_lac_feeding = merge_df.loc[ (merge_df['fluxes_l_plant'] > 0) & (merge_df['fluxes_l_lac'] < 0) ]
```
